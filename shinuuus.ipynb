# Шинус

```python
import math

iterations=39

def sh(x):
    numerator=x
    denominator=1
    n=0
    sum=0
    while n<iterations:
        sum += numerator/denominator 
        numerator *= x**2
        denominator *= (2*n+2)*(2*n+3)
        n+=1
    return sum
```

Легко видеть, что функция чрезмерно верна.

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

angles = np.r_[-3:3:0.001]
plt.plot(angles, np.vectorize(math.sinh)(angles))
plt.plot(angles, np.vectorize(sh)(angles))
plt.show()
```
